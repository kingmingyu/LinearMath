```python
import numpy as np
import matplotlib.pyplot as plt

A=np.array([[7,2],[-7,5]])
b=np.array([-5,12])
x=np.linalg.solve(A,b)
print(x)

x1,y1=x
print(x1,y1)
fig=plt.figure()
ax=fig.add_subplot(1,1,1)

a1=A[:,0]
b1=A[:,1]
c1=-b

for c1,c2,c3 in zip(A[:,0],A[:,1],b):
  x=np.linspace(-7,7,100)
  y=(c3-c1*x)/c2
  ax.plot(x,y,color="black")

ax.plot(x1,y1,'ro')
ax.axis([-7,7,-7,7])
ax.set_xticks(range(-7,7))
ax.set_yticks(range(-7,7))
ax.grid()
ax.set_axisbelow(True)
ax.set_aspect('auto', adjustable='box')
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
a1=np.array([7,-7])
a2=np.array([2,5])
b=np.array([-5,12])

A=np.column_stack((a1,a2))

x=np.linalg.solve(A,b)

x1,y1=x
a1=a1*x1
a2=a2*y1
fig=plt.figure()
ax=fig.add_subplot(1,1,1)

ax.quiver(0,0,a1[0],a1[1],angles='xy',scale_units='xy',scale=1)
ax.text(a1[0]-1,a1[1],'a1',size=15)
ax.quiver(0,0,a2[0],a2[1],angles='xy',scale_units='xy',scale=1)
ax.text(a2[0],a2[1],'a2',size=15)
ax.quiver(0,0,b[0],b[1],angles='xy',scale_units='xy',scale=1,color='blue')
ax.text(b[0],b[1],'b',size=15,color='blue')

ax.set_xticks(range(-10,5))
ax.set_yticks(range(-3,15))
ax.grid()
ax.set_axisbelow(True)
ax.set_aspect('auto',adjustable='box')
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
plt.show()
```

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow


img_path='/content/drive/MyDrive/image.jpg'
img=cv2.imread(img_path)
height, width=img.shape[:2]
M=np.array([[1,0,100],[0,1,200],[0,0,1]],dtype=float)
dst=np.zeros(img.shape,dtype=np.uint8)

for y in range(height-1):
  for x in range(width-1):
    p = np.array([x,y,1])
    p_=np.dot(M,p)

    x_,y_=p_[:2]
    x_=int(x_)
    y_=int(y_)

    if x_>0 and x_<width and y_>0 and y_<height:
      dst[y_,x_] = img[y,x]

result=cv2.hconcat([img,dst])
cv2_imshow( result)
cv2.waitKey(0)
```

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow


img_path='/content/drive/MyDrive/image.jpg'
img=cv2.imread(img_path)
height, width=img.shape[:2]
M=np.array([[1,0,100],[0,1,200]],dtype=float)

dst=cv2.warpAffine(img,M,(width,height))
result = cv2.hconcat([img,dst])
cv2_imshow(result)
cv2.waitKey(0)
```
